```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
df=pd.read_csv("IndiaAffectedWaterQualityAreas.csv",encoding='latin')
```

```python
#checking if the data is loaded by printing first five rows of the dataframe
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#checking for the null values
df.isnull().sum()
```

```python
#exploring the dataset
plt.figure(figsize=(20,10))
x=df['State Name'].value_counts().index
y=df['State Name'].value_counts()
sns.barplot(y,x)
plt.title("State wise Water pollution")
```

FROM THE ABOVE OUTPUT IT IS CLEAR THAT RAJASTHAN IS NUMBER ONE STATE IN TERMS OF WATER POLLUTION FOLLOWED BY BIHAR, ASSAM AND ORRISSA.

```python
plt.figure(figsize=(20,10))
sns.barplot(y=df['District Name'].value_counts().head(20).index,x=df['District Name'].value_counts().head(20).values)
plt.title("Districts With most Polluted Water")
```

FROM THE ABOVE OUTPUT WE CAN SEE THAT BARMER IS THE MOST WATER POLLUTED DISTRICT FOLLOWED BY JODHPUR AND SUPAUL.

```python
plt.figure(figsize=(20,10))
sns.barplot(y=df['Block Name'].value_counts().head(20).index,x=df['Block Name'].value_counts().head(20).values)
plt.title("Blocks With most Polluted Water")
```

ABOVE OUTPUT SHOWS THE BLOCK WISE WATER POLLUTED BLOCKS 

```python
plt.figure(figsize=(20,10))
sns.barplot(y=df['Panchayat Name'].value_counts().head(20).index,x=df['Panchayat Name'].value_counts().head(20).values)
plt.title("Panchayat with most polluted water")
```

```python
plt.figure(figsize=(20,10))
sns.barplot(y=df['Village Name'].value_counts().head(20).index,x=df['Village Name'].value_counts().head(20).values)
plt.title("Villages wise water pollution")
```

```python
plt.figure(figsize=(20,10))
sns.barplot(y=df['Habitation Name'].value_counts().head(20).index,x=df['Habitation Name'].value_counts().head(20).values)
plt.title("Habitation wise water pollution")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y=df['Quality Parameter'].value_counts().index,x=df['Quality Parameter'].value_counts().values)
plt.title("Types of water pollutants and their count")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df['Year'].value_counts().index,y=df['Year'].value_counts().values)
plt.title("Year wise water pollution")
```

```python
#Renaming the columns name
df=df.rename(columns={'State Name':'State','District Name':'District','Block Name':'Block','Panchayat Name':'Panchayat','Village Name':'Village','Habitation Name':'Habitation','Quality Parameter':'Quality'})
```

```python
df.columns
```

```python
plt.figure(figsize=(20,10))
ax= sns.countplot(y=df['Year'], hue=df['Quality'])
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])
ax.bar_label(ax.containers[3])
plt.title('Year wise Distribution of Contamination', size=20)
```

#plotting For Rajasthan

```python
Rajasthan=df[df.State=='RAJASTHAN']
Rajasthan.head()
```

```python
Rajasthan.describe()
```

```python
Rajasthan.District.unique()
```

```python
Rajasthan.info()
```

```python
Rajasthan.Year.unique()
```

```python
Rajasthan['District'].value_counts()
```

```python
Rajasthan['Quality'].groupby(Rajasthan['District']).describe().sort_values(by='count', ascending=False)
```

```python
plt.figure(figsize=(60,80))
ax= sns.countplot(y=Rajasthan['District'], hue=Rajasthan['Quality'])
plt.title('Water Quality Parameter in Different Districts of Rajasthan', size=20)
```

```python
Rajasthan['Year']= pd.to_datetime(Rajasthan['Year'])
Rajasthan['Year']= Rajasthan['Year'].dt.year
```

```python
Rajasthan.head()
```

```python
Rajasthan['Year'].value_counts()
```

```python
plt.figure(figsize=(10,12))
ax= sns.countplot(Rajasthan['Year'], hue=Rajasthan['Quality'])
ax.bar_label(ax.containers[0]);
ax.bar_label(ax.containers[1]);
ax.bar_label(ax.containers[2]);
ax.bar_label(ax.containers[3]);
plt.title('Water Quality Parameter in Rajasthan from 2009 to 2012', size=20);
```

```python
sns.set_context("paper", rc={"font.size":60,"axes.titlesize":60,"axes.labelsize":60})  
sns.set(font_scale = 3)
g= sns.catplot(y="District", col= "Year",data =Rajasthan,kind = "count",
            hue='Quality')
g.set_xticklabels(rotation=90, size=45)
g.fig.set_figwidth(60)
g.fig.set_figheight(30)
g.fig.suptitle('Analysis of Water Quality Parameter in each District by Year',size=100)
plt.show();
```
